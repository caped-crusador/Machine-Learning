## 1.1

In this section, EM with gaussian mixture alorithm is implemented don Iris dataset for 3 clusters. Number of iterations done is 10 and results of cluster assignment is shown on every step of the maximization.

```python
import pandas as pd
from pylab import *
from scipy.stats import norm
from sklearn.metrics import silhouette_score, silhouette_samples

# loading the data
data = np.loadtxt("iris.data", delimiter=",")
sepal_data = data[:, 0:2]
sepal_data = np.insert(sepal_data, 2, data[:, 4], axis=1)
sepal_data = pd.DataFrame(data=sepal_data)

# initial plot
scatter(sepal_data[0], sepal_data[1], c=sepal_data[2], alpha=0.7)
xlabel("sepal length")
ylabel("sepal width")
title("Actual clusters")
legend()
show()
```

```python
# EM function taking number of culsters and iterations as input
# the clusters are set to 2 and iterataions to 10 by default

def EM(clusters=2, iterations=10):
    np.random.seed(5)
    
    # made copy of the data before starting
    sepal_data_copy = sepal_data.copy()
    sepal_data_copy = np.array(sepal_data_copy)

    # defining initial parameters
    mu = sepal_data_copy[np.random.choice(len(sepal_data_copy), clusters, False), 0:2]
    sigmas = [np.eye(2)] * clusters
    likelihoods = [1. / clusters] * clusters
    
    # performing iterations 
    for k in range(iterations):
        for i in range(len(sepal_data_copy)):
            
            # Expectation step
            cl_lls = np.array([])
            for j in range(clusters):
                cl_lls = np.append(cl_lls, log_likelihood([sepal_data_copy[i, 0], sepal_data_copy[i, 1]],
                                                mu[j], sigmas[j], likelihoods[j]))
            sepal_data_copy[i][2] = cl_lls.argmax()
        
        # plots the state of clustes after every iteration
        scatter(sepal_data_copy[:, 0], sepal_data_copy[:, 1], c=sepal_data_copy[:, 2], alpha=0.9)
        xlabel("sepal length")
        ylabel("sepal width")
        title("Clusters after itertaion : %d"%(k+1))
        show()

        # Maximization step 
        point_fractions = np.array([])
        for j in range(clusters):
            point_fractions = np.append(point_fractions,
                                        len(sepal_data_copy[sepal_data_copy[:, 2] == j])/len(sepal_data_copy) )
        likelihoods = point_fractions

        for j in range(clusters):
            mu[j] = np.array([sepal_data_copy[sepal_data_copy[:, 2] == j][:, 0].mean(),
                              sepal_data_copy[sepal_data_copy[:, 2] == j][:, 1].mean()])
            sigmas[j] = np.diag([sepal_data_copy[sepal_data_copy[:, 2] == j][:, 0].std(),
                              sepal_data_copy[sepal_data_copy[:, 2] == j][:, 1].std()])

    return sepal_data_copy, likelihoods

def log_likelihood(datapoint, mu, sig, phi):
    ll = log(phi)
    for i in range(len(datapoint)):
        ll += log(norm.pdf(datapoint[i], mu[i], sig[i][i]))
    return ll

sepal_data_copy, likelihoods = EM(clusters=3, iterations=10)
```

The above program uses log likelihoods instead of the probabilities for better implementation. One class is well defined but the other two classes are mixed ( as seen in the "actual clusters" plot) and are difficult to distinguish.

## 1.2

In this section, we will compare the results of EM algorithm and evaluate the results for different number of clusters. we will be using Silhouette scores and plot to make the comparision for different number of clusters.

###### Silhouette score
Silhouette cluster analysis is the technique that is widely used to determine the optimum number of clusters in a dataset. The silhouette score is a measure that shows that how close each point in one cluster is to the points of its neighbouring cluster.
The range of values of this measure is [-1,1].

```python

def evaluate(clusters):
    y_init = 10
    datapoint_Sil_Vals = silhouette_samples(sepal_data_copy[:, 0:2], sepal_data_copy[:, 2])
    for i in range(clusters):
        cluster_sil_vals = np.sort(datapoint_Sil_Vals[sepal_data_copy[:, 2] == i])
        cluster_size = len(cluster_sil_vals)
        axis_y = y_init + cluster_size

        color = cm.spectral(float(i) / 2)
        fill_betweenx(np.arange(y_init, axis_y),
                      0, cluster_sil_vals, facecolor=color,
                      edgecolor=color, alpha=0.8)
        # Label the silhouette plot
        text(-0.05, y_init + 0.5 * cluster_size, str(i))
        title("Silhoutte plot for %d clusters" % clusters)
        y_init = axis_y + 10
        xlabel("Silhoutte coeff. value")
        ylabel("cluster labels")
    show()
    print("Average Silhoutte score for %d clusters : "%clusters, silhouette_score(sepal_data_copy[:, 0:2], sepal_data_copy[:, 2]))

# ---------------------------------------------------------------------------
evaluate(clusters=3)
```

For three clusters, the average silhoutte score of the plot plot is 0.39. all the clusters found by the algorithm have silhouette scores greater than the average, except for cluster 1 which contains points very close to cluster zero. since initially there are three clusters in the dataset, but two of them are not well defined. this kind of results were expected at some point.

```python
sepal_data_copy, likelihoods = EM(clusters=2, iterations=10)
evaluate(clusters=2)
```

For 2 clusters, the average silhouette score of all the data points is 0.45. Both of the clusters found by teh algorithm have silhoute scores greater than the average. No significant amount of overlapping is there and the clusters seem well defined. So, it may be a good choice for the number of clusters. 

```python
sepal_data_copy, likelihoods = EM(clusters=4, iterations=10)
evaluate(clusters=4)
```

In the case of 4 clusters, the Silhouette average score is 0.34. One cluster is having a value less than that of the average(cluster 2). this means that the algorithm failed to find well defined culsters for the number of clusters given. four clusters for this problen is probably not a good choice.

## 2

In this section, we need to implement a Causal model of the given problemm of car starting. The causal model is made with the lep of python package _lea_ and the desired answers to the questions are calculated and printed out.


```python
from lea import *

# given marginal probabilities
Alternator_broken = Lea.boolProb(1, 1000)
FanBelt_broken = Lea.boolProb(2, 100)

# initially we are assuming the gas and starter both are true
Gas = Lea.boolProb(100, 100)
starter = Lea.boolProb(100, 100)

# likelihood table for battery cahrging or not
Battery_Charging = Lea.buildCPT(
                    (Alternator_broken & FanBelt_broken, Lea.boolProb(0, 1000)),
                    (Alternator_broken & ~FanBelt_broken, Lea.boolProb(0, 1000)),
                    (~Alternator_broken & FanBelt_broken, Lea.boolProb(0, 1000)),
                    (~Alternator_broken & ~FanBelt_broken, Lea.boolProb(995, 1000)))

# Likelihhod table for battery flat or not
Battery_flat = Lea.buildCPT(
                (Battery_Charging, Lea.boolProb(10, 100)),
                (~Battery_Charging, Lea.boolProb(90, 100)))

# Likelihhod table for Car starts or not
Car_starts = Lea.buildCPT((Gas & starter & ~Battery_flat, Lea.boolProb(95, 100)),
                          (Gas & starter & Battery_flat, Lea.boolProb(0, 100)),
                          (Gas & ~starter & ~Battery_flat, Lea.boolProb(0, 100)),
                          (Gas & ~starter & Battery_flat, Lea.boolProb(0, 100)),
                          (~Gas & starter & ~Battery_flat, Lea.boolProb(0, 100)),
                          (~Gas & starter & Battery_flat, Lea.boolProb(0, 100)),
                          (~Gas & ~starter & ~Battery_flat, Lea.boolProb(0, 100)),
                          (~Gas & ~starter & Battery_flat, Lea.boolProb(0, 100)),)


# calculating the desired probabilities
print("2.2 Probability that the Alternator is Broken, given the Car won't start : ",
      Pf(Alternator_broken.given(~Car_starts)), "\n")
print("2.3 Probability that the Fan Belt is Broken, given the Car won't start : ",
      Pf(FanBelt_broken.given(~Car_starts)), "\n")
print("2.4 Probability that the Fan Belt is Broken, given the Car won't start and Alternator is broken : ",
      Pf(FanBelt_broken.given(~Car_starts & Alternator_broken)), "\n")
print("2.5 Probability that the Alternator and the Fan Belt are broken, given the Car won't start : ",
      Pf((Alternator_broken & FanBelt_broken).given(~Car_starts)), "\n")
```

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import confusion_matrix, accuracy_score
from IPython.display import display, HTML
import numpy as np
import pandas as pd
import string
import math

print("---importing the data -----")
categories=['alt.atheism', 'comp.windows.x', 'rec.sport.hockey', 'talk.politics.guns']
newsgroups_train = fetch_20newsgroups(subset="train", remove=('headers', 'footers', 'quotes'),
                                      categories = categories)
train_data = pd.DataFrame({"x": newsgroups_train.data,
                           "y": newsgroups_train.target})

# cleaning the data
def clean_data(x):
    print("---cleaning the data -----")
    x = np.array(x)
    for i in range(len(x)):
        # remove punctuations
        x[i] = x[i].translate(str.maketrans(string.punctuation, ' '*len(string.punctuation)))
        # remove numbers
        x[i] = x[i].translate(str.maketrans("", "", string.digits))
        # convert to lower case
        x[i] = x[i].lower()
    return x
x = clean_data(train_data["x"])

print("--- extracting features ----- ")
# constructing feature vector
vectorizer = CountVectorizer(stop_words="english", binary=True)
X = vectorizer.fit_transform(x)
x = X.toarray()


# splitting into train and test sets
x_train = np.split(x, [int(len(x)*0.7), len(x)])[0]
x_test = np.split(x, [int(len(x)*0.7), len(x)])[1]

y = np.array(train_data["y"])
y_train = np.split(y,[int(len(y)*0.7), len(y)])[0]
y_test = np.split(y,[int(len(y)*0.7), len(y)])[1]


# calculating priors and likelihoods for each word per document class
priors = np.array([])          # p(C=k)
word_likelihoods = np.array([])
for i in np.unique(train_data["y"]):
    print("--- learning class %d ----"%i)
    priors = np.append(priors, len(y_train[y_train == i])/len(y_train))
    if i == 0:
        word_likelihoods = np.append(word_likelihoods,
                                     (np.sum(x_train[y_train == i], axis=0)+1)/(len(y_train[y_train == i])+2))
    elif i == 1:
        word_likelihoods = np.append([word_likelihoods],
                                     [(np.sum(x_train[y_train == i], axis=0)+1)/(len(y_train[y_train == i])+2)],
                                     axis=0)
    else:
        word_likelihoods = np.append(word_likelihoods,
                                     [(np.sum(x_train[y_train == i], axis=0)+1)/(len(y_train[y_train == i])+2)],
                                     axis=0)


# predicting the classes for test data
print("--- Predicting ----")
predicted_class = np.array([])
for j in range(len(x_test)):
    pred_probabs = np.array([])
    for i in np.unique(train_data["y"]):
        posteriors = ((word_likelihoods[i] * x_test[j]) +
                             ((1 - word_likelihoods[i]) * (1 - x_test[j])))
        posteriors = np.prod(posteriors) * priors[i]
        pred_probabs = np.append(pred_probabs, posteriors)

    predicted_class = np.append(predicted_class, np.argmax(pred_probabs))
cfm = confusion_matrix(predicted_class, y_test)
cfm = pd.DataFrame(cfm, columns=categories, index=categories)
print("\nConfusion Matrix : ")
display(cfm)
accuracy = accuracy_score(predicted_class, y_test)
print("\n\naccuracy of the model : %f"%(accuracy*100), "%")
```

The accuracy achieved when classifying text into for classes is around 72%
